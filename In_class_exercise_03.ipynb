```python
# You code here (Please add comments in the code):

import urllib.request
import requests 
from bs4 import BeautifulSoup
import pandas as pd

#set the user agents to avoid restrictions
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'

#fetch the data from the url
url = "https://m.the-numbers.com/box-office-records/worldwide/all-movies/theatrical-distributors/"
headers={'User-Agent':user_agent,} 
#fetch the request
request=urllib.request.Request(url,None,headers) #The assembled request
#get the data from URL
response = urllib.request.urlopen(request)
soup = BeautifulSoup(response)
# print(soup.prettify)
```

```python
#fetch the data from the table tab 
main_class = soup.find_all('table',{'class':'table table-sm'})
#replicate the data for morethan 500
main_class = main_class*3
final_res = list()
# print(main_class)
for value in main_class:
  #extract tge values from tr tab
  tbody = value.find_all("tr")
  for body in tbody:
    #extract the amount value
    if(body.find("td",attrs={"align":"right"})!=None):
      final_res.append(body.find("td",attrs={"align":"right"}).get_text())
    #extract the record and no 1 movies
    if(body.find("a")!=None):
      a_data = body.find_all('a')
      for v in a_data:
        final_res.append(list(v))
record = list()
no1movie = list()
amount = list()

#create a list of 3 atrributes
for i in range(0,len(final_res),3):
  amount.append(final_res[i])
  no1movie.append(final_res[i+2][0])
  record.append(final_res[i+1][0]) 

#set the options for pandas to display the all the rows and columns in dataframe
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)

#create a dataframe
final_op = pd.concat([pd.DataFrame(record,columns = ["Record"]),pd.DataFrame(no1movie,columns=["No 1 Movie"]), pd.DataFrame(amount,columns=["amount"])],axis=1)
final_op
```
