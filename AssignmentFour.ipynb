```
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
from nltk.tokenize import RegexpTokenizer
```

```
data_txs = df["cleaned_reviews"].values
reviews_data1 = [i.split() for i in data_txs]
reviews_data1 = reviews_data1[:500]
reviews1 = reviews_data1[:500]
```

```
bigram1 = gensim.models.Phrases(reviews_data1, min_count=5, threshold=100)
bigram_phaser1 = gensim.models.phrases.Phraser(bigram1)
trigram1 = gensim.models.Phrases(bigram1['vocab'], threshold=100)  
trigram_phaser1 = gensim.models.phrases.Phraser(trigram1)
```

```
def make_bigrams(ip):
    return [bigram_phaser1[t] for t in ip]
def make_trigrams(ip):
    return [trigram_phaser1[bigram_phaser1[t]] for t in ip]
```

```
bigrams1 = make_bigrams(reviews_data1)
trigrams1=make_trigrams(bigrams1)
```

```
id2wrd1 = corpora.Dictionary(trigrams1)
```

```
corpus1 = [id2wrd1.doc2bow(i) for i in trigrams1]
```

```

lda_model1 = gensim.models.ldamodel.LdaModel(corpus=corpus1,
                                           id2word=id2wrd1,
                                           num_topics=20, 
                                           random_state=100)
```

```
res = lda_model1.print_topics()
for i in res:
  print(i)
```
